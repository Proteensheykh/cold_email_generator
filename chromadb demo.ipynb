```python
import chromadb
chroma_client = chromadb.Client()
collection = chroma_client.create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This is a document about New York",
        "This is a document about Lagos"
    ],
    ids=["id1", "id2"]
)
```

```python
all_docs = collection.get(ids='id1')
all_docs
```

```python
results = collection.query(
    query_texts=["This is a query document about corruption"],
    n_results=2
)
results
```
